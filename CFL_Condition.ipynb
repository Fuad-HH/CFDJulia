```julia
function linearconv(nx::Int64)
    l = 2 # length of the domain
    dx = l / (nx-1)
    nt = 20 # number of timesteps
    dt = 0.025 # stepsize
    c = 1 # wave speed in the simple convection equation
    # u_0 is defined as a step function
    u = ones(nx)
    u[convert(Integer, round(0.5/dx)): convert(Integer, round(1/dx + 1))] .= 2;
    un = ones(nt,nx)
    un[1, :] = copy(u)
    for n in 1:nt-1, i in 2:nx
        un[n+1,i] = un[n, i] - c * dt / dx * (un[n, i] - un[n,i-1])
    end


    n=nt
    plot(LinRange(0,l,nx), un[n,:], label="Timestep: $n", ylims=(0.9,2), linecolor = :magenta)
end
```

#### Now create a new linearconv function which is stable.
We will calculate timestep sizes to satisfy the CFL condition

```julia
function linearconv_mod(nx::Int64)
    sigma = 0.5 # our required sigma
    l = 2 # length of the domain
    dx = l / (nx-1)
    dt = sigma * dx # stepsize
    nt = 20 # number of timesteps
    c = 1 # wave speed in the simple convection equation
    # u_0 is defined as a step function
    u = ones(nx)
    u[convert(Integer, round(0.5/dx)): convert(Integer, round(1/dx + 1))] .= 2;
    un = ones(nt,nx)
    un[1, :] = copy(u)
    for n in 1:nt-1, i in 2:nx
        un[n+1,i] = un[n, i] - c * dt / dx * (un[n, i] - un[n,i-1])
    end


    n=nt
    plot(LinRange(0,l,nx), un[n,:], label="Timestep: $n", ylims=(0.9,2), linecolor = :magenta)
end
```

```julia
linearconv_mod(85)
```

```julia
linearconv_mod(41)
```

```julia
linearconv_mod(61)
```
